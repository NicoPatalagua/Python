```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'#Url desde la cuál traeremos nuestro archivo
iris_2d = np.genfromtxt(url, delimiter=',', dtype='object')#Con este método importamos un documento desde la url ya definida. Adémas 
# debemos separar las lineas con un carácter delimitador.
sepallength = iris_2d[:, 0].astype('float')#Realizamos las operaciones con el volumen de datos
petallength = iris_2d[:, 2].astype('float')#Realizamos las operaciones con el volumen de datos
volume = (np.pi * petallength * (sepallength**2))/3#Realizamos la operación con la cuál generaremos datos
volume = volume[:, np.newaxis]#Con el atributo volume agregamos una nueva dimensióón a la mtriz
out = np.hstack([iris_2d, volume])#Adicionamos los valores a la nueva columna
out[:4]#Retornamos la nueva columna
```
